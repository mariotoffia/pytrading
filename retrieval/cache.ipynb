# Sample Use of the Cache Functions

```python
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
import cache
import sqlite3
```

### Cache BTC-USD Ticker

```python
ticker="BTC-USD"
interval="1wk"
start="2023-01-01"
end="2023-12-31"

cache.cache_ticker(ticker=ticker, interval=interval, start=start, end=end, clear=True)
```

### Query the Cached BTC-USD Ticker

```python
db = cache.get_cache_sql_db(ticker, interval)

# Read back and use Date as index column (otherwise pandas will create a new index autoincrement column)
with sqlite3.connect(db) as conn:
    btc_data = pd.read_sql_query("SELECT * FROM ticker_data", conn, index_col='Date')

print(btc_data)
```
